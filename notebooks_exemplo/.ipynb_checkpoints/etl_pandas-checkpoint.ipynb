```python
import pandas as pd
```

```python
base_casas = pd.read_excel('../data/raw/imovel_ribeirao_preto.xlsx',  sheet_name=['Apartamento', 'Casa'])
df_apartamento = base_casas['Apartamento']
df_casa = base_casas['Casa']
```

```python
base_casas_completa = pd.concat([df_apartamento, df_casa])
base_casas_completa.head()
```

```python
base_casas_completa = base_casas_completa[['tipo_imovel', 'url', 'nome', 'preco', 'endereco', 'metragem','quarto', 'banheiro', 'garagem']]
```

```python
base_casas_completa.info()
```

```python
base_casas_completa[['tipo_imovel', 'url', 'endereco']] = base_casas_completa[['tipo_imovel', 'url', 'endereco']].astype('string')
```

```python
base_casas_completa['ID_CASA'] = base_casas_completa['url'].apply(lambda x: x.split('-')[-1].replace('/', '')).astype('int')
base_casas_completa = base_casas_completa.query('preco != "Sob Consulta"')
```

```python
base_casas_completa['preco'].unique()
```

```python
base_casas_completa['preco'] = base_casas_completa['preco'].str.replace('\nPreço abaixo do mercado', '')
base_casas_completa['preco'] = base_casas_completa['preco'].str.replace('R$', '')
base_casas_completa['preco'] = base_casas_completa['preco'].str.strip()
```

```python
base_casas_completa
```

```python
base_casas_completa['preco'] = base_casas_completa['preco'].str.replace('.', '').astype(float)
```

```python
base_casas_completa['endereco_split'] = base_casas_completa['endereco'].apply(lambda x : x.split(','))
```

```python
base_casas_completa
```

```python
base_casas_completa['total_enderecos'] = base_casas_completa['endereco_split'].apply(len)
```

```python
base_casas_completa.head()
```

```python
base_casas_completa['bairro_teste'] = base_casas_completa.apply(
    lambda x: x['endereco_split'][2].split('-')[1]
    if x['total_enderecos'] == 4 
    else ( 
         x['endereco_split'][1].split('-')[-1] if x['total_enderecos'] == 3  
      
        else  x['endereco_split']
    )
    , axis=1
).astype('string')
```

```python
base_casas_completa[['bairro_teste', 'endereco_split', 'total_enderecos']].query('total_enderecos == 2')
```

```python
for bairro in base_casas_completa['bairro_teste'].unique():
    print(bairro)
```

```python
base_casas_completa['bairro_flag'] =  base_casas_completa.apply(
    lambda x : 1 if isinstance(x['bairro_teste'], str) and '[' in x['bairro_teste'] else 0, axis=1)
base_casas_completa[['bairro_flag', 'bairro_teste']]
```

```python
base_casas_completa['bairro_teste'] = base_casas_completa.apply(
    lambda x: x['bairro_teste'].replace("['", "").replace("']", "") if x['bairro_flag'] == 1 else x['bairro_teste'], axis=1
)
```

```python
base_casas_completa['bairro_teste'] = base_casas_completa.apply(
    lambda x: x['bairro_teste'] if '-' not in bairro 
    else ( 
        x['bairro_teste'].split('-')[1].split(',')[0] 
        if(len(x['bairro_teste'].split('-')) == 3) 
        else (x['bairro_teste'].split('-')[2].split(',')[0] if len(x['bairro_teste'].split('-') )== 4 else x['bairro_teste'].split('-')[0].split(',')[0])
        )
        , axis=1
)
```

```python
base_casas_completa['bairro_teste'] = base_casas_completa['bairro_teste'].str.replace('"',"").str.strip().str.title().str.replace("'", "").astype('string')
```

```python
for bairro in base_casas_completa['bairro_teste'].unique():
    print(bairro)

 
```
